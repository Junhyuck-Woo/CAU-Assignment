# import library

```python
import numpy as np; import matplotlib.pyplot as plt
```

# Define a differentiable function that maps from real number to real numbe
$f(x){\;}={\;}{x^3}{\;}-{\;}{4x}$

# Define a domain of the function.
$-10{\;} {\leqq}{\;}X{\;}{\leqq}{\;}10$

```python
x = np.arange(-10, 10, 0.1); y = x*x*x - 40*x
```

# Plot the function.

```python
plt.figure(figsize=(5, 3));plt.plot(x, y, color='blue'); 
plt.title('Taylor Approximation'); plt.xlabel('x'); plt.ylabel('y');
```

# Select a point within the domain.

```python
p = 7; m = p*p*p - 40*p
```

# Mark the selected point on the function.

```python
plt.figure(figsize=(5, 3));plt.plot(x, y, color='blue'); plt.plot(p, m, 'ro');
plt.title('Taylor Approximation'); plt.ylabel('y');plt.xlabel('x'); 
```

# Define the first-order Taylor approximation at the selected point.
${ \hat{f}{(x)}{\;}={f(z)\;} + {\partial f \over\partial x}{(z)}{(x - z)}}$

```python
fh = (p*p*p - 40*p) + (3*p*p - 40)*(x - p) 
```

# Plot the Taylor approximation with the same domain of the original function.

```python
plt.figure(figsize=(5, 3));plt.plot(x, y, color='blue'); plt.plot(x, fh, color='green'); plt.plot(p, m, 'ro'); 
plt.title('Taylor Approximation');plt.xlabel('x'); plt.ylabel('y');
```
